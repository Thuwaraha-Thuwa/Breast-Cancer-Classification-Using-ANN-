```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
x= data.drop(columns ='lable',axis =1)
y =data['lable']
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=2)
#random state mean it will select data for traing and testing randomly 
```

```python
#Standardize the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python


#fit_transform:

#fit: Calculates the mean and standard deviation of each feature in X_train.
#transform: Applies the standardization formula to scale the features:
#𝑧=𝑥−𝜇/𝜎
'''transform (only):
Uses the same mean and standard deviation calculated from X_train (from the fit step).
Standardizes X_test based on these values.
This ensures that the testing data is transformed consistently with the training data.'''
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)

x_test_std = scaler.transform(x_test)
```

```python
'''Random operations (e.g., initializing weights, shuffling data, or creating random numbers) are often used in machine learning models.
By setting a specific seed value (in this case, 3), the sequence of random numbers generated becomes deterministic. This means every time you run your code, you will get the same random values, ensuring consistent results.'''

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras






```

```python
#set the modeles and layer to nural network
from tensorflow.keras import models,layers
models= models.Sequential([
     layers.Dense(32,input_shape=(30,),activation='sigmoid'),
     layers.Dense(64,activation='relu'),
     layers.Dense(32,activation='relu'),
     layers.Dense(2,activation='sigmoid')
])
```

```python
#compile the nueron network
models.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
print(y_train)
```

```python
#traing the neural network
models.fit(x_train_std,y_train,validation_split=0.1,epochs=20)
```

```python
#accuracy of the model on test data
loss, accuracy = models.evaluate(x_test_std, y_test)
print(accuracy)


```
